```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np


# Create a simple dataset
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)

# Applying DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)

# Visualizing the clusters
plt.figure(figsize=(8, 6))

# Core samples in black
core_samples_mask = np.zeros_like(dbscan.labels_, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
plt.scatter(X[core_samples_mask, 0], X[core_samples_mask, 1], c='k', marker='o', s=100, label='Core Samples')

# All samples colored by cluster
unique_labels = set(clusters)
for label in unique_labels:
    if label == -1:
        col = 'gray'  # Noise points in gray
    else:
        col = plt.cm.nipy_spectral(label / len(unique_labels))
    class_member_mask = (clusters == label)
    xy = X[class_member_mask & core_samples_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=col, edgecolor='k', s=50, label=f'Cluster {label}')
plt.title('DBSCAN Clustering')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Create a simple dataset
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Applying Hierarchical Clustering
agg_cluster = AgglomerativeClustering(n_clusters=4, linkage='ward')
clusters = agg_cluster.fit_predict(X)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', edgecolor='k')
plt.title('Hierarchical Clustering')
plt.show()
```
